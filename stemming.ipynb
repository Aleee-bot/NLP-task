```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer , RegexpStemmer
```

```python
words = ["run", "rqunning", "runs", "write", "writing", "eats", "eaten", "programming", "programs"]
```

```python
stemming = PorterStemmer()
for word in words:
    print(word + "--->" + stemming.stem(word))
```

```python
# ing$ --> removes "ing" from the end of the word
# ing --> removes "ing" from the everywher on the word
# min = 4 --> only removes suffix with minimum length of 4 characters

reg_stemmer = RegexpStemmer("ing|s$|es$|ed$|able$" , min = 4)
wordss = [ "his", "eats", "eatable" , "ingeating", "writing","learned","goes"]

for word in wordss:
    print(word + "--->" + reg_stemmer.stem(word))
```

```python
snowball_stemmer = SnowballStemmer("english")
for word in words:
    print(word + "--->" + snowball_stemmer.stem(word))
```

```python
#difference between porter and snowball stemmer is that snowball stemmer is more aggressive than porter stemmer and it also takes care of the irregular forms of the words.
print(f"Snowball Stemmer output : {snowball_stemmer.stem("fairly")}, {snowball_stemmer.stem("sportingly")}")
print(f"Porter Stemmer output : {stemming.stem("fairly")}, {stemming.stem("sportingly")}")
```
