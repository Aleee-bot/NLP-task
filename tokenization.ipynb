```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus = """Hello world , how are you?
This is a NLP task.
I need $4.25."""
```

```python
print(corpus)
```

```python
#sent_tokenize() breaks text into individual sentences and returns as a list.
from nltk.tokenize import sent_tokenize
document = sent_tokenize(corpus)
print(document)
len(document)
```

```python
#word_tokenize() breaks text into individual words and punctuation marks as separate tokens
from nltk.tokenize import word_tokenize
tokens = word_tokenize(corpus)
print(tokens)
len(tokens)
```

```python
#wordpunct_tokenize() breaks text into individual words and separates ALL punctuation marks as individual tokens.
from nltk.tokenize import wordpunct_tokenize
documentss = wordpunct_tokenize(corpus)
print(documentss)
len(documentss)
```

```python
#TreebankWordTokenizer() uses the Penn Treebank tokenization rules to break text into words and punctuation.
#It is a sophisticated tokenizer.
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
toks = tokenizer.tokenize(corpus)
print(toks)
len(toks)
```
