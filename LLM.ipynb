```python
dataset_imdb = load_dataset("stanfordnlp/imdb")
```

```python
dataset_imdb
```

```python
dataset_imdb['test']['text']
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenize_datasets = dataset_imdb.map(tokenize_function, batched=True)
tokenize_datasets
```

```python
small_train_dataset = tokenize_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenize_datasets["test"].shuffle(seed=42).select(range(1000))
```

```python
from transformers import TrainingArguments

training_arg = TrainingArguments(
    output_dir = "./results",
    eval_strategy = "epoch",                     
    learning_rate = 2e-5,
    per_device_train_batch_size = 16,
    per_device_eval_batch_size = 16,
    num_train_epochs = 3,
    weight_decay = 0.01
    )
```

```python
from transformers import AutoModelForSequenceClassification  , Trainer
model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-uncased" , num_labels = 2)
```

```python
Training = Trainer(
    model = model ,
    args = training_arg ,
    train_dataset = small_train_dataset ,
    eval_dataset = small_eval_dataset
)
Training.train()
```

```python
result = Training.evaluate()
result

Training.save_model("./resultt")
```
