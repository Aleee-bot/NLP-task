```python
paragraph = "The children are playing in the gardens and running around the houses happily. One child ran faster than the others and was shouting loudly. Their parents were watching them carefully and were talking about how the kids have grown over the years. A teacher is teaching lessons in the nearby school, and she teaches very interesting subjects. The boys have completed their assignments, but one student was still working on his project. Cars are moving quickly on the roads, and drivers were driving more carefully than usual because it was raining heavily."
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentence = nltk.sent_tokenize(paragraph)

sentence
```

```python
for i in range(len(sentence)):
    words = nltk.word_tokenize(sentence[i])
    parag = []
    for word in words:
        if word.lower() not in set(stopwords.words('english')):
            parag.append(word)
    pos_tagg = nltk.pos_tag(parag)

    print(pos_tagg)
```
