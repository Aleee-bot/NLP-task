```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
```

```python
#POS tag for verb is 'v' and for noun is 'n' and for adjective is 'a' and for adverb is 'r'
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("running" , pos = 'n')
```

```python
words = ["run", "running", "runs", "write", "writing", "eats", "eaten", "programming", "programs"]
for word in words:
    print(word + "-->" + lemmatizer.lemmatize(word , pos = 'v'))
```
